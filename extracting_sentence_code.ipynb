```python
from numpy import random
import numpy as np
import pandas as pd
```

```python
removal = ['Full text:', 'illustration photo:', 'illustration caption:', 'photo:',
            'photographer:', 'photo (color):', 'caption:', 'credit:', 'styling:']
small_endings = ['mr.','mrs.','miss.','ms.','am.','pm.', 'inc.', 'st.','rd.','ave.', 'r.',
                'sr.' ,'a.m.','p.m.','a.m','p.m', 'n.']
EOL = re.compile(r"([.!?]{1,}[\s]+)|(\.\")|(\.\')")
SPACES = re.compile(r"[\s]{2,}")
brackets = re.compile(r"\([A-Za-z0-9: ]+\)?")
symbols = re.compile(r'[^A-Za-z ]')
dashes = re.compile(r'([\w]+([\s]+)?[\-]{1,}([\s\-]+)?[\w]+)')
n = []

for f in files:
    with open(folder+'/'+f, encoding = 'utf-8') as text:
        all_lines = []
        lines = text.readlines()
        lines = ' '.join(lines)
        lines = lines.replace('\n', '. ').replace('. .', '.')\
                    .replace('\r', '').lower()
        for x in removal:
            lines = lines.replace(x, '')
        lines = re.sub(r'[\.]{2,}','.', lines)
        lines = lines.replace('___','').replace('__','')

        for x in small_endings:
            lines = re.sub(r'[^\w]({})[^\w]'.format(x), ' '+x[:-1]+' ', lines)
        lines = re.sub(r'[\s]{2,}', ' ', lines)

        # 1955. for mr.
        # 1. mel gibson
        # Listed stuff

#         rep = [x for x in re.findall(r'[\d]+[\.]+[\s]+?[A-Za-z0-9\s]{2,}[\.]', lines)
#                if x.count(' ') <= 3]
#         for x in rep:
#             lines = lines.replace(x, x.replace('.',''))

        #beauties = set(re.findall(r'beaut[\w]+', lines))
        beauties = {'beauty'}

        lines = re.split(EOL, lines)
        lines = [x for x in lines if x is not None]
        i = 0
        while i < len(lines)-1:
            if len(lines[i+1].replace(' ','')) <= 2:
                lines[i] += lines[i+1]
            i += 1

        lines = [re.sub(r'[\s]{2,}',' ', x).lstrip(' ').rstrip(' ') for x in lines]
        lines = [x for x in lines if x.count(' ') > 1 and len(x) > 3]

        all_words = []
        for x in lines:
            for i in beauties:
                if i in x:
                    all_words.append(x)
                    break

        all_words = np.array(all_words)
        indices = list(range(len(all_words)))
#         random.shuffle(indices)
#         first_ten = int(len(indices)/10*2)
#         indices = indices[:first_ten]
#         indices = sorted(indices)
#         all_words = all_words[indices]

    text = pd.Series(all_words)
    text.index = indices
    text.to_csv(f'{folder}/{f}(final).csv')
    n.append(len(text))
```
